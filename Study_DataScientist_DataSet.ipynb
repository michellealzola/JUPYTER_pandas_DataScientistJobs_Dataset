```python
# Loading and Getting Information
```

```python
df = pd.read_csv('DataScientist.csv', index_col='Job Title')
```

```python
df
```

```python
df_data = df.drop(['Unnamed: 0', 'index'], axis=1)
```

```python
df_data
```

```python
# Problem 1: Which top 15 companies are hiring the most data scientists?
```

```python
top_companies_hiring_ds = df_data['Company Name'].value_counts().nlargest(15)
```

```python
top_companies_hiring_ds
```

```python
top_companies_hiring_ds.plot(kind='bar')
plt.title('Top 15 Companies that are hiring Data Scientists')
plt.xlabel('Company')
plt.ylabel('Job Postings')
plt.show()
```

```python
# Problem 3: What is the average salary for data scientists per sector?
```

```python
df_data.info()
```

```python
df_data['Salary Estimate'] = df_data['Salary Estimate'].apply(lambda x: x.split('(Glassdoor est.)'))
```

```python
df_data['Salary Estimate']
```

```python
df_data['Salary Estimate'] = df_data['Salary Estimate'].apply(lambda  x: x[0])
```

```python
df_data['Salary Estimate']
```

```python
v1 = []
for values in df_data['Salary Estimate']:
    v1.append(values.replace('K', '000'))

```

```python
v1
```

```python
v2 = []
for values in v1:
    v2.append(values.replace('$', ''))
```

```python
v2
```

```python
v3 = []
for values in v2:
    v3.append(values.split('-'))
```

```python
v3
```

```python
v4 = []
for values in v3:
    int1 = int(values[0])
    v4.append(int1)
```

```python
v4
```

```python
df_data['Salary Estimate'] = v4
```

```python
df_data
```

```python
df_sector_average_salary = df_data.groupby('Sector')['Salary Estimate'].mean()
```

```python
df_sector_average_salary
```

```python
df_sector_average_salary.plot(kind='bar')
plt.title('Average Salary per Sector based on Minimum Salary Offered')
plt.show()
```

```python
# Problem 4: How many data scientist jobs require specific skills?
```

```python
skills = {'Python': 0,
          'R': 0,
          'SQL': 0,
          'Excel': 0,
          'Machine Learning': 0,
          'Deep Learning': 0,
          'NLP': 0,
          'Tableau': 0
}
```

```python
for index, row in df_data.iterrows():
    for skill in skills.keys():
        if skill in row['Job Description']:
            skills[skill] += 1
```

```python
print(skills.keys(), skills.values())
```

```python
plt.bar(skills.keys(), skills.values())
plt.xticks(rotation=90)
plt.title('Specific Skills required for a Data Scientist Job based on Job Postings')
plt.xlabel('Job Postings')
plt.show()
```
